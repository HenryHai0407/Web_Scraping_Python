### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]
```

```python
plt.plot(x,y)
plt.title('Plot Example')
plt.xlabel('X Value')
plt.ylabel('Y Value')
plt.show()
```

```python
import numpy as np
x = np.linspace(0, 10, 1000)
```

```python
plt.plot(x,np.sin(x), color='blue',linestyle='dashed')
plt.xlabel('X')
plt.ylabel('Y value')
plt.axis([0,4,-0.8,0.8])
```

```python
plt.plot(x,np.sin(x),color='blue',linestyle='dashed',label='Sin(X)')
plt.plot(x,np.cos(x),color="Red",label='Cos(X)')
plt.axis('equal')
plt.legend()
```

```python
 months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
 sales_product_a = [200, 250, 300, 350, 400, 450]
 sales_product_b = [180, 220, 290, 340, 390, 460]
```

```python
plt.style.use('seaborn-v0_8-darkgrid')
plt.plot(months, sales_product_a, label='Product A', marker='o')
plt.plot(months, sales_product_b, label='Product B', marker='s')
plt.legend()
```

```python
# Bar chart
plt.bar(months, sales_product_a, width=0.4, label='Product A', align='center')
plt.bar(months, sales_product_b, width=0.4, label='Product B', align='edge')
plt.legend()
```
