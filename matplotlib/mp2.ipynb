```python
df = sns.load_dataset('tips')
df
```

```python
# Create a scatter plot
plot = sns.scatterplot(x='total_bill', y='tip', hue='sex', data=df)
plot.set_title('The correlation between Total Bill and Tip')
plot.set_xlabel('Total Bill')
plot.set_ylabel('Tip')
```

```python
grouped_data = df.groupby('sex', as_index=False)['total_bill'].sum()
sns.barplot(x='sex', y='total_bill', data=grouped_data)
```

```python
grouped_data1= df.groupby('smoker',as_index=False)['total_bill'].sum()
sns.barplot(x='smoker',y='total_bill',data=grouped_data1)
```

```python
# Heatmap chart
import numpy as np
sales_data = np.array([
    [200, 250, 300, 350, 400, 450],
    [180, 220, 290, 340, 390, 460],
    [150, 200, 240, 300, 350, 400]
 ])
```

```python
# Draw heatmap
sns.heatmap(sales_data, annot=True, cmap="YlGnBu", xticklabels=['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun'], yticklabels=['Product A', 'Product B', 'Product C'])
```
