```python
import pandas as pd
```

```python
file_path = 'D:/repos/scraping_data_from_web/pandas/data.csv'
df = pd.read_csv(file_path)
```

```python
df
```

```python
df.rename(columns={'Age':'Customer_Age'}, inplace = True)
```

```python
df.replace({'Gender': {'M':'Male','F':'Female'}}, inplace = True)
```

```python
# Apply function
def age_div(age):
    return age / 2

df['Age_div_def'] = df['Customer_Age'].apply(age_div)
```

```python
print(df)
```

```python
# df = df0.copy() if necessary, to avoid modifying df0
df.loc[df['Sales'] < 200, 'Sales'] = df['Sales'] /2
```

```python
print(df)
```

```python
df[df['Name'].str.contains('a')]
```

```python
# Creating Unique_ID column for df
df['Unique_ID'] = df['CustomerID'].astype(str) + df['Name'].str[:3] + df['Customer_Age'].astype(str).str[:1]
```

```python
df
```

```python
df2 = df.copy()
```

```python
df['Sales'].astype(str).str.len()
```

```python
# Work with DateTime
df2['JoinDate'] = pd.to_datetime(df2['JoinDate'])
df2['Year'] = df2['JoinDate'].dt.year
df2['Month'] = df2['JoinDate'].dt.month
df2.info()
```

```python
# Use Timedelta to add 5 days more to the current JoinDate
df2['JoinDate+5'] = df2['JoinDate'] + pd.Timedelta(days=5)
df2
```

```python
# Create Sales Category
df2['Sales_Category'] = df2['Sales'].map(lambda x: 'High Sales' if x > 350 else 'Low Sales')
df2
```

```python
# Drop unnecessary columns
df3 = df2.drop(columns=['Age_div_def','JoinDate+5','Unique_ID','Year'])
```

```python
df3
```

```python
# Fill NA in Customer_Age column with mean value
df3['Customer_Age'] = df3['Customer_Age'].fillna(df3['Customer_Age'].mean())
df3['Customer_Age'] = df3['Customer_Age'].apply(lambda x: int(x))
df3
```

```python
# Group by Gender, Region, Month
df3.groupby('Gender')['Sales'].sum()
```

```python
df3.groupby('Region')['Sales'].sum()
```

```python
# Group by 'Month' column to calculate total sales in descending order
df3.groupby('Month')['Sales'].sum().sort_values(ascending=False)
```
