```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = {
'Name': ['John', 'Mary', 'DAVID', 'Emily', 'Michael', 'Sarah', np.nan, 'Lisa'],
'Age': [25, 31, 42, 28, 35, 'twenty-two', 38, 45],
'Country': ['USA', 'Canada', 'UK', 'Australia', 'Germany', 'France', 'Italy', 'spain'],
'Salary': [50000, 60000, 70000, 80000, 90000, 100000, 110000, 120000],
'Email': ['john@example.com', 'mary@example.com', 'david@example.com', 'emily@example', np.nan,
'sarah@example.com', 'tom@example.com', 'lisa@example.com']
}
```

```python
data = pd.DataFrame(data)
data.head()
```

```python
data.replace('DAVID','David',inplace=True)
data
```

```python
data.fillna('No Information',inplace=True)
data
```

```python
data['Age'].replace('twenty-two',22,inplace=True)
```

```python
data['Salary'].clip(lower=50000,upper=120000)
data
```

```python
data['Country'] = data['Country'].mask(~data['Country'].str.istitle())
data
```
