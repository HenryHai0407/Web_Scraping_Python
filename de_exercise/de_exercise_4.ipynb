#### Inspecting DataFrames in Pandas

```python
data.head()
```

```python
data[data['Sector'] == 'Health Care'].head(10)
```

```python
# Using Sector-based mean imputation
data['Dividend Yield'] = data.groupby('Sector')['Dividend Yield'].transform(
    lambda x: x.fillna(x.mean())
)
data['Price/Earnings'] = data.groupby('Sector')['Price/Earnings'].transform(
    lambda x: x.fillna(x.mean())
)
```

```python
data.info()
```

```python
data.sample()
```

```python
data['Sector'].value_counts()
```

```python
data[data['Sector'] == 'Health Care'].filter(items=['Symbol','Name','Sector','Price']).head()
```

```python
hc_data = data[data['Sector'] == 'Health Care'][['Symbol','Name','Sector','Price']]
hc_data.head()
```

```python
# 2/ Basic calculation:
data['P/E to Dividend'] = data['Price/Earnings']/data['Dividend Yield']
data.head()
```

```python
# Mean, Min, Max of P/E to Dividend column
data['P/E to Dividend'].describe()
```

```python
data.head()
```

```python
# 3/ Sector analysis
data.groupby('Sector')['Market Cap'].sum()
```

```python
# Clean Sector column
data['Sector'] = data['Sector'].str.strip()
data['Sector'] = data['Sector'].str.title()
data['Sector'] = data['Sector'].replace({
    'Industries': 'Industrials'
})
```

```python
sector_marketcap = data.groupby('Sector')['Market Cap'].sum()
sector_marketcap
```

```python
# Plot a bar chart
sector_marketcap.plot(kind='bar', color='skyblue')
plt.title('The Market Cap for each sector')
plt.xlabel('Sector')
plt.ylabel('Market Cap')
```

```python
# 4/ Top 10 biggest market-cap companies
top10_marketcap = data.sort_values(by='Market Cap',ascending=False).head(10)
top10_marketcap
```

```python
# Plot a bar chart
top10_marketcap.plot(kind='bar', x='Name', y='Market Cap', color='skyblue')
plt.title('Top 10 highest Market-Cap companies')
plt.xlabel('Companies')
plt.ylabel('Market Cap')
```

```python
# 5/ Correlation between P/E and Dividend Yield
data.plot.scatter(x='Price/Earnings',y='Dividend Yield')
```

Comment: No clear correlation between Price/Earnings and Dividend Yield.

```python
# 6/ 52 week range:
data['52 week range'] = data['52 week high'] - data['52 week low']
```

```python
data['52 week range'].describe()
```

```python
# Company who has the Min value of 52 week range
data[data['52 week range'] == data['52 week range'].min()]
```

```python
# Company who has the Max value of 52 week range
data[data['52 week range'] == data['52 week range'].max()]
```

```python
# 7/ Rank companies based on financial figure (Price/Sales)

data['rank'] = data['Price/Sales'].rank(method='first')
data.sort_values(by='rank',ascending=True).head()
```

```python
top5_companies = data.sort_values(by='rank',ascending=False).head(5)
top5_companies
```

```python
# 8/ Financial ratios by industry:
avg_ratios = data.groupby('Sector')[['Price/Book','Price/Sales']].mean()
avg_ratios
```

```python
# Plot avg_pb chart
avg_ratios.plot(kind='bar',figsize=(12,6))
plt.title('The average of Price/Book and Price/Sales by Sector')
plt.ylabel('Values')
plt.xlabel('Sector')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.legend(title='Financial ratio')
```

```python
# 9/ Histogram of P/E and EBITDA:
# P/E histogram:
plt.figure(figsize=(8,4))
plt.hist(data['Price/Earnings'].dropna(),bins=10, color='skyblue', edgecolor='black')
plt.title('The histogram of Price/Earnings')
plt.xlabel('P/E ratio')
plt.ylabel('Frequency')
```

```python
# P/E Histogram (Zoom In):
pe_data = data['Price/Earnings']
pe_filtered = pe_data[(pe_data > 0) & (pe_data < 200)]

plt.figure(figsize=(8, 4))
plt.hist(pe_filtered, bins=15, color='skyblue', edgecolor='black')
plt.title('The histogram of Price/Earnings from 0 to 200')
plt.xlabel('P/E Ratio')
plt.ylabel('Frequency')
```

```python
# EBITDA histogram:
plt.figure(figsize=(8,4))
plt.hist(data['EBITDA'].dropna(),bins=10, color='lightgreen', edgecolor='black')
plt.title('The histogram of EBITDA')
plt.xlabel('EBITDA')
plt.ylabel('Frequency')
```

```python
# EBITDA histogram (Zoom In):
eb_data = data['EBITDA']
eb_filtered = eb_data[(eb_data > 0) & (eb_data < 15)]

plt.figure(figsize=(8, 4))
plt.hist(eb_filtered, bins=15, color='lightgreen', edgecolor='black')
plt.title('The histogram of EBITDA from 0 to 15')
plt.xlabel('EBITDA')
plt.ylabel('Frequency')
```

```python
# 10/ Heatmap analysis:
import seaborn as sns
metrics = ['Price', 'Dividend Yield', 'Price/Earnings', 'Earnings/Share',
           'Book Value', 'Market Cap', 'EBITDA', 'Price/Sales', 'Price/Book']

financial_data = data[metrics].dropna()
```

```python
correlation_matrix = financial_data.corr()
correlation_matrix
```

```python
# Plot a heatmap
plt.figure(figsize =(10,8))
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('The correlation matrix of financial metrics')
```

Comment:

 *Strong correlation pairs*: 

        EBITDA vs. Market Cap
    
        Price vs. Earnings/Share
