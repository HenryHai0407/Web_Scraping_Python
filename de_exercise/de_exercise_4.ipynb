## Matplotlib with Pandas

```python
import matplotlib.pyplot as plt
```

```python
hc_data.head(5).plot(kind='bar',x='Name',y='Price',color='skyblue')
plt.title('Price of Each Company in SP500')
plt.xlabel('Name')
plt.ylabel('Price')
plt.show()
```

```python
data['Sector'].value_counts().plot(kind='pie', autopct = '%1.1f%%')
```
